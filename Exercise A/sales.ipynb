## Numerical analysis and visualization

We'll analyze the Unit_Cost column: -

```python
sales['Unit_Cost'].describe()
```

```python
sales['Unit_Cost'].mean()
```

```python
sales['Unit_Cost'].plot(kind='box',vert=True, figsize=(7,7))
```

```python
sales['Unit_Cost'].plot(kind='density',figsize=(7,7))
```

```python
ax = sales['Unit_Cost'].plot(kind='density', figsize=(7,7))
ax.axvline(sales['Unit_Cost'].mean(),color='purple')
ax.axvline(sales['Unit_Cost'].median(),color='green')
```

```python
ax = sales['Unit_Cost'].plot(kind='hist',figsize=(7,7))
ax.set_ylabel('Number of Sales')
ax.set_xlabel('dollars')
```

## Relationship between the columns

Can we find any significat relationship from dataframe

```python
sales_numeric = sales.apply(pd.to_numeric, errors='coerce')

corr = sales_numeric.corr()
corr
```

```python
fig = plt.figure(figsize=(7,7))
plt.matshow(corr, cmap='Accent', fignum = fig.number)
plt.xticks(range(len(corr.columns)), corr.columns,rotation='vertical');
plt.yticks(range(len(corr.columns)),corr.columns);
```

```python
sales.plot(kind='scatter', x ='Customer_Age' ,y='Revenue',figsize=(7,7))
```

```python
ax = sales[['Profit','Age_Group']].boxplot(by='Age_Group',figsize=(7,7))
ax.set_ylabel('Profit')
```
